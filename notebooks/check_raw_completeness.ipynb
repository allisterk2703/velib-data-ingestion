##### Check that every "raw" day has 96 files

```python
RAW_DATA_PATH = Path("../data/station_status/raw")

for year in sorted(os.listdir(RAW_DATA_PATH)):
	for month in sorted(os.listdir(RAW_DATA_PATH / year)):
		for file in sorted(os.listdir(RAW_DATA_PATH / year / month)):
			if len(os.listdir(RAW_DATA_PATH / year / month / file)) != 96:
				print(f"‚ùå {RAW_DATA_PATH / year / month / file}: {len(os.listdir(RAW_DATA_PATH / year / month / file))} files")				
```
